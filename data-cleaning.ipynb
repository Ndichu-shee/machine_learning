```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
health_pd = pd.read_csv('./Downloads/health/2022/heart_2022_with_nans.csv')
```

```python
health_pd.head(5)
```

```python
health_pd.duplicated().value_counts()
```

```python
health_pd=health_pd.drop(columns=['RemovedTeeth','HadStroke','HadSkinCancer','HadAngina','HadCOPD','HadDepressiveDisorder','DeafOrHardOfHearing','DifficultyConcentrating','HadArthritis','HadDiabetes','LastCheckupTime','HIVTesting','FluVaxLast12','PneumoVaxEver','BlindOrVisionDifficulty','TetanusLast10Tdap','CovidPos','ECigaretteUsage','SmokerStatus','RaceEthnicityCategory','DifficultyWalking','DifficultyDressingBathing','DifficultyErrands'])
```

```python
health_pd.fillna({'PhysicalHealthDays': 0.0, 'MentalHealthDays': 0.0,'SleepHours':0.0}, inplace=True)
```

```python
health_pd['WeightInKilograms'] = health_pd['WeightInKilograms'].interpolate(method='linear', limit_direction='backward')
```

```python
health_pd['WeightInKilograms'].isnull().value_counts()
```

```python
height_mode = health_pd['HeightInMeters'].mode().round(1)[0]
```

```python
health_pd['HeightInMeters'] = health_pd['HeightInMeters'].fillna(height_mode)
```

```python
health_pd['HeightInMeters'].isnull().value_counts()
```

```python
bmi_mean = health_pd['BMI'].mean()
```

```python
bmi_mean = round(bmi_mean, 1)
```

```python
health_pd['BMI'] = health_pd['BMI'].fillna(bmi_mean)
```

```python
health_pd['BMI'].isnull().value_counts()
```

```python
health_pd['WeightInKilograms'].isnull().value_counts()
```

```python
health_pd.fillna({'AlcoholDrinkers': 'Unknown', 'AgeCategory': 'Unknown','GeneralHealth':'Unknown','HighRiskLastYear':'Unknown','ChestScan':'Unknown'}, inplace=True)
```
